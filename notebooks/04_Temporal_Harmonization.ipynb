```python
import sys
from pathlib import Path

sys.path.append(
    str(Path.cwd().parent / "src")
)

from extraction.gee import initialize

from preprocessing.loader import load_all
from preprocessing.cleaner import clean_all

from extraction.sites import get_site
from extraction.products import get_product
from extraction.extractor import extract_timeseries

from harmonization.temporal import align_to_common_dates
```

```python
initialize()

DATA_DIR = (
    Path.cwd().parent
    / "data"
    / "raw"
    / "bharatflux"
)
```

```python
datasets = load_all(DATA_DIR)

cleaned_datasets, validation_summary = clean_all(
    datasets
)
```

```python
observed = cleaned_datasets[
    "BFT_2016_LE_ET_dmean"
].data

observed.head()
```

```python
site = get_site("BFT")
product = get_product("MOD16A2GF")

satellite = extract_timeseries(
    site,
    product,
    "2016-01-01",
    "2016-12-31",
)

satellite.head()
```

```python
observed_aligned, satellite_aligned = (
    align_to_common_dates(
        observed,
        satellite,
    )
)
```

```python
print("Observed:", observed.shape)
print("Satellite:", satellite.shape)

print("Observed Aligned:", observed_aligned.shape)
print("Satellite Aligned:", satellite_aligned.shape)
```

```python
observed_aligned["DoY"].equals(
    satellite_aligned["DoY"]
)
```
