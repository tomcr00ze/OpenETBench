```python
import sys
from pathlib import Path

sys.path.append(
    str(Path.cwd().parent / "src")
)

# --------------------------------------------------
# BharatFlux
# --------------------------------------------------

from preprocessing.loader import load_all
from preprocessing.cleaner import clean_all

# --------------------------------------------------
# Earth Engine
# --------------------------------------------------

from extraction.gee import initialize
from extraction.sites import get_site
from extraction.products import get_product
from extraction.extractor import extract_timeseries

# --------------------------------------------------
# Harmonization
# --------------------------------------------------

from harmonization.temporal import (
    align_to_common_dates,
)

from harmonization.merge import (
    merge_observed_satellite,
)

# --------------------------------------------------
# Benchmarking
# --------------------------------------------------

from benchmarking.metrics import (
    calculate_metrics,
)

# --------------------------------------------------
# Visualization
# --------------------------------------------------

from visualization.scatter import (
    plot_scatter,
)
from visualization.maps import (
    plot_site,
)
```

```python
initialize()

DATA_DIR = (
    Path.cwd().parent
    / "data"
    / "raw"
    / "bharatflux"
)

FIGURE_DIR = (
    Path.cwd().parent
    / "figures"
)
```

```python
datasets = load_all(DATA_DIR)
cleaned_datasets, validation_summary = clean_all(
    datasets
)
observed = cleaned_datasets[
    "BFT_2016_LE_ET_dmean"
].data

observed.head()
```

```python
site = get_site("BFT")

product = get_product("MOD16A2GF")

satellite = extract_timeseries(
    site,
    product,
    "2016-01-01",
    "2016-12-31",
)

satellite.head()
```

```python
observed_aligned, satellite_aligned = (
    align_to_common_dates(
        observed,
        satellite,
    )
)
merged = merge_observed_satellite(
    observed_aligned,
    satellite_aligned,
)

merged.head()
```

```python
report = calculate_metrics(
    merged
)
report
```

```python
figure_path = plot_scatter(
    merged=merged,
    metrics=report,
    product_name="MOD16A2GF",
    site="BFT",
    year=2016,
    output_dir=FIGURE_DIR,
)

print(figure_path)
```

```python
from IPython.display import Image

Image(filename=figure_path)
```
